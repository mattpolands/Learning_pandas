```python
#desire multiple conditions 
#using 'and' confuses the system since 'and' relays strictly true and false 
#using 'and' here returns an 'ambiguous' error essentially saying the system is too confused
#using parentheses and the ampersand helps pandas differentiate the request
df[(df['Column A']>0) & (df['Column C']>1)]
#output is the only row that meets both of these conditions
```

```python
#when want to use the "or" operator the same confusion can happen
#for pandas in this situation we use the pipeline '|' operator
df[(df['Column A']>0) | (df['Column C']>1)]
```

```python
#close parentheses in order to run desired method
#can set this parameter to permanently change df by calling "inplace" parameter in the method and changing it to =true
df.reset_index()
```

```python
#creating index levels in pandas
#zip creates a set of tuples
#MultiIndex is a function used to make a customization in a dataframe
outside = ['chair', 'chair', 'chair', 'table', 'table', 'table']
inside = [1,2,3,1,2,3]
hierarchy_index = list(zip(outside, inside))
hierarchy_index = pd.MultiIndex.from_tuples(hierarchy_index)
```

```python
hierarchy_index
```

```python
df3 = pd.DataFrame(randn(6,2),hierarchy_index,['A','B'])
```

```python
df3
```

```python
#indexing off multi-level
df3.loc['chair']
```

```python
#then can index deeper into that index
df3.loc['chair'].loc[2]
```

```python
#output is basically saying there are two indexes and neither has a name
df3.index.names
```

```python
df3.index.names = ['Furniture', 'Number']
```

```python
df3
```

```python
#indexing and pulling a single value
df3.loc['table'].loc[2]['B']
```

```python
#cross section (xs) method allows us a little more flexibility than .loc 
#particularly useful when attempting to pull out elements from two separate indexes
df3.xs(1,level='Number')
```
